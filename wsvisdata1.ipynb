```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_csv=pd.read_csv('/mnt/MyDrive/DMUPB/dataset/Global_Superstore2.csv',encoding='latin-1')
data_csv.info()
```

```python
data_xls=pd.read_excel('/mnt/MyDrive/DMUPB/dataset/Global_Superstore2.xlsx')
data_xls.info()
```

```python
data_xls.head()
```

```python
data_xls['Country'].value_counts()
```

```python
data_xls['Country'].unique()
```

```python
market_ina=data_xls[data_xls['Country']=='Indonesia']
market_ina.head()
```

```python
cols=['City','State','Sales','Product Name','Quantity','Market','Region']
market_ina[cols]
```

```python
market_ina.State.unique()
```

```python
#Jumlah sales setiap state
sales_state=market_ina.groupby('State')['Sales'].sum().sort_values(ascending=True)
sales_state
```

```python
plt.xticks(rotation='vertical')
plt.barh(sales_state.index,sales_state);
```

```python
plt.hist(market_ina['Sales'],bins=20);
```

```python
plt.pie(sales_state.head(10),labels=sales_state.head(10).index);
```

```python
plt.boxplot(sales_state);
```
